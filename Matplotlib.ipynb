### Matplotlib

Matplotlib is a plotting library for python programming language and it's numerical mathematics extension Numpy. It provides an object-oriented API for embedding plots into applicaions using general-purpose GUI toolkits Tkinte, wxPython, QT or GTK+.

Some of the major props of Matplotlib are:
* Generally easy to get started for simple plots
* Support for ustom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python

## Simple examples
x = np.arange(0,10)
y=np.arange(11,21)
```

### Creating a plot to show a sine wave form

```python
np.pi
```

```python
# Compute x and y cordinates for points on a sine curve
x = np.arange(0, 4* np.pi,0.1)
y = np.sin(x)

plt.title('Sine wave form')

# plot the points using matplotlib
plt.plot(x,y);
```

```python
# Subplots
# Compute the x and y cordinates for points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos=np.cos(x)

# Setup a subplot grid that has height 2 and width 1
# and set the first subplot as active.
plt.subplot(2,1,1)

# make the first plot
plt.plot(x,y_sin,'g--')
plt.title('Sine')

# Set the second subplot as active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('Cosine')

# Show the figure
plt.show()
```

### Bar Plot

```python
x = [2,8,10]
y = [11,16,9]

x2=[3,9,11]
y2= [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color='g')

plt.title('Bar Graph')
plt.ylabel('y-axis')
plt.xlabel('x-axis')

plt.show()
```

### Histograms

```python
a=np.array([22,87,5,15,3,32,36,61,66,46,66,1,3,123,161,])
plt.hist(a, bins=20) # bins paramter is deafult 10, but we can increase it if required
plt.title('Histogram')
plt.show()
```

### Box Plots

```python
data = [np.random.normal(0,std,100) for std in range(1,4)] # range is 1-3, 0 to std with 100 steps

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True); # vert parameter is for vertical boxes and patch_artist is for color
```












### Pie Chat

```python
# Data to plot
labels = 'Python','C++','Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
# explode 1st slice, spilits the slice away from pie chart
# As we gave 4 values in sizes, pie chart will have 4 slices gave 0.1 value in explode for first slice
explode= (0.1,0,0,0)  

# plot
plt.pie(sizes,explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%',shadow=True) # autopct is for floating values and shadow is for shadow of the chart

plt.axis('equal')
plt.show();
```
