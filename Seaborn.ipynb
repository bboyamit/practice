### Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very iportant  in Feature Selection.

```python
# Since Correlation only works with numbers, it will not work with objects
# So I have picked number columns only
a =df[['total_bill','tip','size']].corr()
a
```

```python
# visualizing the corr() with sns.heatmap()
sns.heatmap(a);
```

### Jointplot
A Join plot allows to study the relationsip between two numeric variables. The central chart displays their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or 2D density plot.

### Univariate Analysis

```python
# using hexagon
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex');
```

```python
# using regression line
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg');
```

### Pairplot
A pairplot is also known as a scatterplot, in which one variable in the same data row is mathced with another variable's value. Pair plots are just elaboratrs, showing all variables paired with all the other variables

```python
# To ignore warnings
import warnings
warnings.filterwarnings('ignore')

# plot pairplots
sns.pairplot(df);
```

```python
# We can add a filter (legend) 
sns.pairplot(df,hue='sex')
```

### Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
# it will show density
sns.distplot(df['tip'])
```

```python
# to see the distributions by count, turn off kde with kde=False parameter
sns.distplot(df['tip'],kde=False, bins=10)
```

### Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data Points. 
* boxplot
* violinplot
* countplot
* bar plot


### Count plot

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

## Bar plot

```python
sns.barplot(x='sex',y='tip', data=df)
```

### Box Plot

```python
sns.boxplot(x='sex',y = 'tip', data=df)
```

```python
sns.boxplot(x='day', y= 'total_bill', data=df, palette='rainbow')
```

```python
# boxplot on entire data will pick all features with numeric value
sns.boxplot(df,orient='v') # orient to select h for horizontal ad v for vertical
```

```python
sns.boxplot(y='total_bill', x='day',hue='smoker', data=df)
```

### Violin plot
Violin plot helps us to see both the distribution in items of Kernel density estimation and the box plot

```python
sns.violinplot(y='day', x='total_bill', data=df, palette='rainbow')
```
